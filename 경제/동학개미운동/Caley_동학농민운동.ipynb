 비유의 원개념인 동학농민운동은 조선 제 26대 고종 대 1894년 1월 10일부터 4월 27일, 그리고 같은 해 9월 총 2번 일어난 한국 역사 최초의 시민혁명이다. 이 운동은 외세에 맞서고, 평등사상을 바탕으로 봉건질서를 무너뜨리려 했다. 동시에 제도적으로 계급을 철폐를 꾀했다. 반외세, 반봉건의 자주 정치조직이었다는데 의의가 있다.
